```python
import random  # Import the random module to randomly select responses

# Define a mock (fake) LLM class
class NakliLLM:

    # Constructor method: called when an instance of NakliLLM is created
    def __init__(self):
        print('LLM created')  # Print a message indicating that the LLM has been initialized

    # predict method: takes a user prompt and returns a response
    def predict(self, prompt):
        """
        Args:
            prompt (str): The input text or question from the user

        Returns:
            dict: A dictionary containing a randomly selected response
        """

        # Predefined list of example responses
        response_list = [
            'Delhi is the capital of India',
            'IPL is a cricket league',
            'AI stands for Artificial Intelligence'
        ]

        # Randomly select one response from the list and return it in a dictionary
        return {'response': random.choice(response_list)}
```

```python
# A mock/fake prompt template class
class NakliPromptTemplate:

    # Constructor: initialize the template and input variables
    def __init__(self, template, input_variables):
        """
        Args:
            template (str): A string template with placeholders for variables
            input_variables (list of str): List of variable names that the template expects
        """
        self.template = template              # Store the template string
        self.input_variables = input_variables  # Store the list of expected variables

    # Method to fill the template with actual values
    def format(self, input_dict):
        """
        Args:
            input_dict (dict): A dictionary mapping variable names to their values

        Returns:
            str: The formatted string with variables replaced by their values
        """
        # Use Python string formatting to replace placeholders with actual values
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}',
    input_variables=['length', 'topic']
)
```

```python
prompt = template.format({'length':'short','topic':'india'})
prompt
```

```python
llm = NakliLLM()
```

```python
llm.predict(prompt)
```

```python
#connect the prompt and llm together
# A mock/fake LLM Chain class that combines a prompt template with an LLM
class NakliLLMChain:

    # Constructor: initialize with an LLM instance and a prompt template
    def __init__(self, llm, prompt):
        """
        Args:
            llm: An instance of a mock or real LLM class with a `predict()` method
            prompt: An instance of NakliPromptTemplate or similar prompt template
        """
        self.llm = llm        # Store the LLM instance
        self.prompt = prompt  # Store the prompt template instance

    # Method to run the chain: format the prompt and get a response from LLM
    def run(self, input_dict):
        """
        Args:
            input_dict (dict): Dictionary of input variables for the prompt

        Returns:
            str: The response from the LLM after formatting the prompt
        """

        # 1. Fill the prompt template with actual input values
        final_prompt = self.prompt.format(input_dict)

        # 2. Pass the formatted prompt to the LLM's predict method
        result = self.llm.predict(final_prompt)

        # 3. Extract the 'response' from the returned dictionary
        return result['response']
```

```python
template = NakliPromptTemplate(
    template='Write a {length} poem about {topic}',
    input_variables=['length', 'topic']
)
```

```python
llm = NakliLLM()
```

```python
chain = NakliLLMChain(llm, template)
```

```python
chain.run({'length':'short', 'topic': 'india'})
```
